```python
import pandas as pd
import ast
```

```python
results_df = pd.read_csv('./clean/results_jo.csv')
hosts_df = pd.read_csv('./clean/hosts.csv')
athletes_df = pd.read_csv('./clean/athletes.csv')
```

```python
results_df.loc[(results_df.duplicated())]
results_df = results_df.rename({'slug_game':'game_slug'},axis=1)
results_df.columns
```

```python
jo_types = hosts_df[['game_slug','game_season', 'game_year']]
hosts_results_merge = results_df.merge(jo_types, on='game_slug')
hosts_results_merge.head()
```

```python
summer_games_results = hosts_results_merge.loc[(hosts_results_merge.game_season == 'Summer')].copy()
winter_games_results = hosts_results_merge.loc[(hosts_results_merge.game_season == 'Winter')].copy()
summer_games_results.head()
```

```python
summer_games_results.drop(['game_season'], axis=1, inplace=True)
summer_games_results.head()
```

```python
summer_games_results['medal_type'] = summer_games_results['medal_type'].fillna('None')
summer_games_results['total_medals'] = summer_games_results['medal_type'].apply(lambda x: 0 if x == 'None' else 1)
summer_games_results['gold_medals'] = summer_games_results['medal_type'].apply(lambda x: 1 if x == 'GOLD' else 0)
summer_games_results['silver_medals'] = summer_games_results['medal_type'].apply(lambda x: 1 if x == 'SILVER' else 0)
summer_games_results['bronze_medals'] = summer_games_results['medal_type'].apply(lambda x: 1 if x == 'BRONZE' else 0)
summer_games_results.head()
```

```python
summer_games_results['participant_type'].unique()
summer_games_results_individual = summer_games_results.loc[(summer_games_results.participant_type == 'Athlete')].copy()
summer_games_results_individual.head()
```

```python
# def extract_athlete_name(athletes_str):
#     if not pd.isna(athletes_str):
#         athletes_list = ast.literal_eval(athletes_str)
#         if athletes_list and isinstance(athletes_list, list):
#             return athletes_list[0][0] # Retourne le nom de l'athlète
#     return ""

# def extract_athlete_name(athletes_str):
#     if isinstance(athletes_str, str):
#         athletes_list = ast.literal_eval(athletes_str)
#         if athletes_list and isinstance(athletes_list, list):
#             return athletes_list[0][0] # Retourne le nom de l'athlète
#     return ""

def extract_athlete_name(athletes_str):
    if isinstance(athletes_str, str):
        try:
            athletes_list = ast.literal_eval(athletes_str)
            if athletes_list and isinstance(athletes_list, list):
                return athletes_list[0][0]  # Return the name of the first athlete
        except (ValueError, SyntaxError):
            return ""
    return ""

def extract_athlete_url(athletes_str):
    if isinstance(athletes_str, str):
        try:
            athletes_list = ast.literal_eval(athletes_str)
            if athletes_list and isinstance(athletes_list, list):
                return athletes_list[0][1]  # Return the name of the first athlete
        except (ValueError, SyntaxError):
            return ""
    return ""
```

```python
from sklearn.model_selection import train_test_split

df = summer_games_results_individual

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

features = ['actual_age', 'athlete_full_name', 'country_name', 'total_medals_all_time', 'discipline_title', 'event_title']
target = 'total_medals_all_time'
```

```python
X_train = train_df[features]
y_train = train_df[target]
X_test = test_df[features]
y_test = test_df[target]
```

```python
numeric_features = ['actual_age', 'total_medals_all_time']
categorical_features = ['country_name', 'discipline_title', 'event_title']
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
numeric_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42))
])

# Train the model
model.fit(X_train, y_train)

# Evaluate the model
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
